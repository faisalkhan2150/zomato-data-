# Which resturant are more rated in each city .

```python
k=india.groupby(["City","Restaurant Name","Aggregate rating"])["Votes"].agg(["mean"]).reset_index()
k.rename(columns={"mean":"Votes"},inplace=True)
k
```

```python
k.sort_values(by=["Aggregate rating","Votes"],ascending=False)
```

# Find best restaurants in New Delhi .

```python
k[k["City"]=="New Delhi"].sort_values(by=["Aggregate rating","Votes"],ascending=False)
```
